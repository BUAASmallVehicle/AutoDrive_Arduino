本方案首先通过全局路径生成一系列的路径点，每两个路径点之间的部分使用A\*算法进行寻路.

## 与简单A\*算法的区别
在实际道路中使用的A\*算法，与上一部分有所不同，它更加复杂一些. 

**第一**，在实际道路中，不可能遍历图中的每一个像素来进行路径搜寻，这样计算成本非常大. 因此我们将地图(或者说是路面)划分成一系列矩形区域，在这些区域中分别取点，将这些点作为A\*算法中搜寻的点，如图所示：

![](实际道路_选点.png)

这样可以减少计算成本，使得小车能够及时给出路径决策.

**第二**，在实际路况下，成本函数需要重新定义，$g(n)$与$h(n)$并不能很好地给出路径的成本. 事实上，在实际中需要考虑的成本包括偏离成本(偏离车道中心线的程度)、碰撞成本(远离障碍物的程度)、舒适成本(加速度不宜过快)与法律成本(是否违章)等，因此在需要定义一个更为复杂的成本函数.

**第三**，在寻路时，使用Fernet坐标系(即s-q坐标系)更为方便，因此我们需要将原先的直角坐标系抛弃，采用Fernet坐标.

**第四**，同(1)，为了能够及时做出路径决策，我们需要将从起点至终点的路径划分成若干段，在每一段内进行路劲规划，做出决策，下一次需要决策的区段以本次终点为起点. 

**第五**, 在小车项目中，地图是通过摄像获取的，在一切开始之前，需要将摄像头捕获的图像进行处理，转变成能够使用的地图.

**我们希望构建地图的过程能够在识别过程中就实现，这样就不需要再单独做一个模块来构建地图，既浪费计算资源，又耽误时间**

```python
# 为了之后的研究中方便将图像转化成所需要的地图格式，构造一个小函数来方便标记路面
def road_surface(img_thresh):
    for x in range(img_thresh.shape[1]):
        where = 'out_road'
        for y in range(img_thresh.shape[0]):
            # 如果标签显示尚未进入车道
            if where == 'out_road':
                # 如果已到达路边线
                if img_thresh[y][x] == 255:
                    where = 'in_top_side_line'
                continue
            # 如果在车道上边线上
            elif where == 'in_top_side_line':
                # 如果已经到达路面
                if img_thresh[y][x] == 0:
                    where = 'in_line'
                    img_thresh[y][x] = 84
                continue
            # 如果已经在路面上
            elif where == 'in_line':
                # 如果未触到边线
                if img_thresh[y][x] == 0:
                    img_thresh[y][x] = 84
                # 如果触到边线
                elif img_thresh[y][x] == 255:
                    where = 'in_bottom_side_line'
            # 如果在车道下边线上
            elif where == 'in_bottom_side_line':
                continue
        
    return img_thresh

# 测试一下
thresh_to_map_test = thresh1.copy()
# 去障与中心分道线
barrier_1 = np.array([[70, 160], [100, 130]])
barrier_2 = np.array([[330, 420], [30, 80]])
center_line = np.array([[0, 681], [85, 100]])
thresh_to_map_test[barrier_1[1][0]:barrier_1[1][1], barrier_1[0][0]:barrier_1[0][1]] = 0
thresh_to_map_test[barrier_2[1][0]:barrier_2[1][1], barrier_2[0][0]:barrier_2[0][1]] = 0
thresh_to_map_test[center_line[1][0]:center_line[1][1], center_line[0][0]:center_line[0][1]] = 0

# 输出结果
plt.imshow(road_surface(thresh_to_map_test))
plt.show()
        
```

## 2 - 直角坐标与Fernet坐标转换
我们需要将直角坐标系转换成Fernet坐标系，这样更为方便.

### 2.1 - 寻点、拟合基准线.
首先要找一些点来作为拟合点.

```python
# 为了抓重点，暂时不考虑选点策略，直接采用中心点来作为拟合点
fitting_pts = []
for x in range(thresh_curveRoad.shape[1]):
    # 每隔30个像素取一个点
    if ((x + 1) % 30) == 0:
        # 找到中点
        p = list(np.where(thresh_curveRoad[:, x] == 255)[0])
        if p != []:
            a = min(p)
            b = max(p)
            c = int((a + b)/2)
            fitting_pts.append([c, x])
        
print(fitting_pts)
```

```python
np.array(fitting_pts).T
```

```python
plt.scatter(np.array(fitting_pts).T[1], np.array(fitting_pts).T[0])
```

```python
# 定义自变量x与因变量y
x = np.array(fitting_pts).T[1]
y = np.array(fitting_pts).T[0]

# 经过观察，本例中四次拟合效果更好
fitting_curve3 = np.polyfit(x, y, 4)
fitting_function3 = np.poly1d(fitting_curve3)
print(fitting_function3)
```

```python
# 绘制出拟合曲线
yval = np.polyval(fitting_function3, x)

plot1 = plt.plot(x, y, 's',label='original values')
plot2 = plt.plot(x, yval, 'r',label='polyfit values')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=4) #指定legend的位置右下角
plt.title('polyfitting')
plt.show()
```
