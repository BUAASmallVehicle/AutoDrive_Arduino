# 车道线保持

如何得到车道线保持信息，这是本项目中需要自行解决的问题. 根据查阅到的论文资料，为了保持小车能够沿着车道线行驶，需要知道当前道路车道线的近似曲线，根据近似曲线求出切线方向，让车始终沿着切线方向行驶. 

目前车道线保持算法的思路**暂定**为：
1. 从lane_detection模块获取提取出的车道线信息，并进行透视变换；
2. 确定车道线上的点，由这些点构建车道线的近似曲线作为基准；
3. 确定当前车辆位置前的某个位置，计算该处基准线的切线方向(让切点位置靠前，主要是为了避免运算和通信导致的延迟)；
4. 将切线方向参数传递给vehicle_control模块，控制车辆的转向角度.

如图，是一张棋盘，工作者进行了定位，确定了棋盘的四个角的在图片中的位置：[[129, 118], [179, 119], [78, 216], [219, 216]]

之后根据棋盘的实际大小，对转换后的图像四个角进行了定位，得到转换后的位置：[[120, 180], [261, 180], [120, 431], [261, 431]]

接下来就是进行透视变换：

```python
pt1_tr = [110, 70]
pt2_tr = [115, 170]
pt3_tr = [215, 70]
pt4_tr = [215, 170]

# 输入转换前的坐标与转换后的坐标
pts1 = np.float32([[129, 118], [179, 119], [78, 216], [219, 216]])
pts2 = np.float32([[120, 180], [261, 180], [120, 431], [261, 431]])

M = cv2.getPerspectiveTransform(pts1, pts2)


dst = cv2.warpPerspective(img_tran_org, M, (400, 500))


plt.imshow(img_tran_org), plt.title('Input')
plt.show()
plt.imshow(dst), plt.title('Output')
plt.show()
```

上图就是原图与经过变换后的图片.

```python
img_tran_org = cv2.imread('/home/lhospital/MyProgramm/AutoDrive_Arduino/FrameWork/version_model/车道保持研究/变换矩阵_原图.jpeg')
img_tran_obj = cv2.imread('/home/lhospital/MyProgramm/AutoDrive_Arduino/FrameWork/version_model/车道保持研究/变换矩阵_变换.jpeg')

print(img_tran_org.shape)
# print(img_tran_obj.shape)

img_tran_org = imresize(img_tran_org, (120, 160, 3))
img_tran_obj = imresize(img_tran_obj, (160, 120, 3))


plt.imshow(img_tran_org)
plt.show()
plt.imshow(img_tran_obj)
plt.show()
```

在拍摄时，左右两个哑铃片之间的距离为半条车道，约1.75米，前后两个哑铃片之间的距离大约半条白线，约3.75米.

由原图可以看出，底部两个哑铃片的坐标分别为：\[[70, 115], [113, 115]\]，每米在图片上占$(113 - 70)/1.75$个像素.
顶部两个哑铃片的坐标分别为: \[[69, 110], [99, 110]\]. 

经过计算，顶部两个哑铃片在新的图片中的坐标为: \[[70, 115 - 3.73*(113 - 70)/1.75], [113, 115 - 3.73*(113 - 70)/1.75]\]

```python
# 输入转换前的坐标与转换后的坐标
pts1 = np.float32([[69, 110], [99, 110], [70, 115], [113, 115]])
pts2 = np.float32([[200, 115 - 3.73*(113 - 70)/1.75], [341, 115 - 3.73*(113 - 70)/1.75], [200, 431], [341, 431]])

M = cv2.getPerspectiveTransform(pts1, pts2)


dst_img_tran = cv2.warpPerspective(img_tran_org, M, (450, 700))

plt.imshow(img_tran_org), plt.title('Input')
plt.show()
plt.imshow(dst_img_tran), plt.title('Output')
plt.show()
```

上图已经较好地将原图进行反透视，但仍有缺陷，因为原图中更远的信息并没有展示出来.
